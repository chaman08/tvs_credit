**Data Cleaning**

```python
raw_data.isnull().sum()
```

```python
data = raw_data.drop(['BUREAU_SCORE','MATCH_SCORE','PAYNIMO_STATUS','DISBMONTH','LOAN_NUMBER','CUST_PINCODE'],axis = 1)
#drpoping BUREAU_SCORE,MATCH_SCORE,PAYNIMO_STATUS bacause it has missing values more than 5%
#dropping DISBMONTH and LOAN_NUMBER because creating dummy for this does not make the sense
#dropping CUST_PINCODE because we already creating dummy for state to take geography as input, taking each pincode will create lots of dummy wil cause loosing of model efficiency
data.head()
```

```python
data.isnull().sum()
```

```python
data_cleaned = data.dropna()
#now we are safe to remove empty rows because we are removing less than 5% of data
data_cleaned.head()
```

```python
data_cleaned.isnull().sum()
#now we have no missing values
```

```python
#Let's check descritive of cleaned data
data_cleaned.describe()
```

```python
data_cleaned.info()
```

it seems like we are good to go further

**Dealing with non numerical data**
by creating dummies and mapping

```python
data_cleaned['CD_SEGMENT'] = data_cleaned['CD_SEGMENT'].map({'MOBILE':1,'Non-Mobile':0})
data_cleaned['ETC_NTC'] = data_cleaned['ETC_NTC'].map({'ETC':1,'NTC':0})
data_cleaned['GENDER'] = data_cleaned['GENDER'].map({'M':1,'F':0})
data_cleaned['MARITALSTATUS'] = data_cleaned['MARITALSTATUS'].map({'Married':1,'Single':0,'N':0})
data_cleaned.head()
```

```python
data_with_dummies = pd.get_dummies(data_cleaned,drop_first = True, dtype=int)
data_with_dummies.head()
```

```python
data_with_dummies.info()
```

**Variable**

```python
data_with_dummies.columns.values
```

```python
x1 = data_with_dummies.drop(columns = 'TARGET_VARIABLE', axis = 1)
y = data_with_dummies['TARGET_VARIABLE']
```

**Regression**

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
model = LogisticRegression()
```

```python
# training the Logistic Regression Model
model.fit(x1, y)
```

```python
coefficients = model.coef_[0]  # Coefficients for the features
coefficients
```

```python
intercept = model.intercept_[0]  # Intercept term
intercept
```

```python
# accuracy
x1 = model.predict(x1)
data_accuracy = accuracy_score(x1, y)
print('Accuracy on data : ', data_accuracy)
```

```python
import statsmodels.api as sm
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x1)
results_log = reg_log.fit()
results_log.summary()
```
